```python
import numpy as np
import matplotlib.pyplot as plt
x_train = np.array([1.0,2.0])
y_train = np.array([300.0, 500.0])
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
print(f"x_train.shape = {x_train.shape}")

#m - number of training examples

m = x_train.shape[0]
print(f"Number of training examples is: {m}")
```

```python
i = 0 #Phần tử đầu tiên
x_i = x_train[i]
y_i = y_train[i]
print(f"(x^({i})), y^({i}) = ({x_i}, {y_i})")
```

```python
#plot data
plt.scatter(x_train, y_train, marker = 'x', c = 'r')
plt.title("Bảng giá nhà dựa vào diện tích nhà")
plt.ylabel('Giá (theo 1000$)')
plt.xlabel('Diện tích nhà ($ft^2$)')
plt.show()
```

```python
w = 200
b = 100
print(f"w: {w}")
print(f"b: {b}")
```

```python
def compute_cal_func(x,w,b):
  """Với x (ndarry(m,)): Data với m examples
      w: vector hệ số
      b: bias của model
      f_wb = y^: Dự đoán của mô hình"""
  m = x.shape[0]
  f_wb = np.zeros(m) #Mảng 1 chiều
  for i in range(m):
    f_wb[i] = w * x[i] + b

  return f_wb
```

```python
twr_func = compute_cal_func(x_train, w,b)
#Plot của mô hình dự đoán
plt.plot(x_train, twr_func, c = 'b', label = 'Giá dự đoán')

#plot của các điểm data
plt.scatter(x_train, y_train, marker = 'x', c = 'r', label = 'Giá trị thực tế')

plt.title("Giá nhà")
plt.xlabel("Diện tích nhà (theo thang 1000ft$^2$)")
plt.ylabel("Giá nhà (theo thang 1000$)")
plt.legend()
plt.show()
```
